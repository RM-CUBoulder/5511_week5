```python
# Read necessary modules
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
from os import listdir
from os.path import isfile, join
from itertools import chain
from PIL import Image
import random
```

```python
# Function to compute length of files 
def filelength(string, a):
    mypath = string
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    print("The folder with {} contains {} files".format(a,len(onlyfiles)))    
```

```python
# Function to plot file sizes of all files in a directory
def plot_filesize(filepath, kaggle):
    size_list = []
    mother = filepath #'/kaggle/input/gan-getting-started/monet_jpg/'

    mypath = mother
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

    for i in range(len(onlyfiles)):
        size_list.append(os.path.getsize(mother+ onlyfiles[i])) 

    plt.hist(size_list, edgecolor= 'white')
    plt.xlabel('Bytes')
    plt.ylabel('Frequency')
    plt.title('File size folder {} in bytes'.format(mother[34:]))
    plt.savefig(kaggle)
```

```python
# Plotting of file size images in monet_jpg folder
file_size_monet = plot_filesize('/kaggle/input/gan-getting-started/monet_jpg/', '/kaggle/working/monet.png')
file_size_monet
```

## Plot RGB

```python
# Function to plot RGB dimensions of a .jpg image 
def get_dimensions(filepath):
    
    # List all files in a directory
    size_list = []
    mother = filepath 

    mypath = mother
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

    # Randomly select file from directory
    selected_item = random.choice(onlyfiles)

    # Image in .jpg format
    im =  Image.open(mother + selected_item)

    # Convert image to array
    im_array = np.array(im)

    ## Select RGB elements from array
    # Red 
    R_list = list(chain.from_iterable(im_array[:,:,0]))
    # Green
    G_list = list(chain.from_iterable(im_array[:,:,1]))
    # Blue
    B_list = list(chain.from_iterable(im_array[:,:,2]))

    # Plot histograms for the RGB elements
    fig, ax = plt.subplots(1, 3, figsize=(10, 4))

    print(selected_item)

    ax[0].hist(R_list, color = 'red', edgecolor='white', alpha = 0.4)
    ax[0].set_ylabel('Frequency')
    ax[0].set_xlabel('Degree of Red')
    ax[0].set_ylim(0,30000)

    ax[1].hist(G_list, color = 'green', edgecolor='white', alpha = 0.4)
    ax[1].set_ylabel('Frequency')
    ax[1].set_xlabel('Degree of Green')
    ax[1].set_title('RGB dimensions {}'.format(selected_item), size = 16, pad = 16)
    ax[1].set_ylim(0,30000)

    ax[2].hist(B_list, color = 'blue', edgecolor='white', alpha = 0.4)
    ax[2].set_ylabel('Frequency')
    ax[2].set_xlabel('Degree of Blue')
    ax[2].set_ylim(0,30000)

    fig.tight_layout()
    
    return im
```
