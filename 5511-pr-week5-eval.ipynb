```python
# Read necessary modules
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
from os import listdir
from os.path import isfile, join
from itertools import chain
from PIL import Image
import random
```

```python
# Function to plot randomly chosen generated images.
def eval_plot(transform):
    # Set seed so the three pictures are chosen.
    random.seed(5)
    # Make a list of integers 0 through 7.037
    int_list = list(np.arange(0,7038))
    # Select three random integers from int_list
    int_list_sample = random.sample(int_list, 3)
    # The file path of the images 
    comp = transform
    # Get all the complete file paths (including the images)
    mypath = comp
    compfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    # Set up the plotting so pictures are shown horizontally
    _, ax = plt.subplots(1,3, figsize=(12, 12))
    # In a loop select the images corresponding to the integers in int_list_sample 
    for i in range(3):
        pic = compfiles[int_list[i]]
        
        monet = Image.open(comp + pic)

        ax[i].imshow(monet)
        ax[i].axis("off")
    
    plt.savefig('/kaggle/working/Arch.jpg') 
```

```python
# Plot transformed images of architecture 1 
eval_plot('/kaggle/input/transform/images/')
```

```python
# Plot transformed images of architecture 2
eval_plot('/kaggle/input/transform/images_I/')
```

```python
# Plot transformed images of architecture 3
eval_plot('/kaggle/input/transform/images_II/')
```
